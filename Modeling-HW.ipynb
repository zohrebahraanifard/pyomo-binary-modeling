* #####  Importing pyomo

```python
import pyomo.environ as pyo
```

* #### Constructing Model

```python
model= pyo.ConcreteModel()
```

* #### Sets

```python
model.Iset = pyo.RangeSet(1,8)
model.Jset = pyo.RangeSet(1,8)
model.K1Set= pyo.RangeSet(-7,7)
model.K2Set= pyo.RangeSet(2,16)
```

* #### Variable

```python
model.delta= pyo.Var(model.Iset, model.Jset , domain= pyo.Binary)
```

* #### Objective Function

```python
def obj_rule(model):
    return sum(model.delta[i,j] for i in model.Iset for j in model.Jset)
model.obj = pyo.Objective( rule= obj_rule , sense=pyo.maximize)
```

* #### Constraints

* #### Part A

```python
def const1_rule(model, i):
    return sum(model.delta[i,j] for j in model.Jset)<=1
model.const1 = pyo.Constraint(model.Iset, rule=const1_rule)
```

```python
def const2_rule(model, j):
    return sum(model.delta[i,j] for i in model.Iset)<=1
model.const2 = pyo.Constraint(model.Jset, rule= const2_rule)
```

* #### Part B

```python
def const3_rule(model, k):
    return sum(model.delta[i,j] for i in model.Iset for j in model.Jset if i-j==k)<=1
model.const3= pyo.Constraint(model.K1Set, rule = const3_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part B
model.const3.deactivate()
```

```python
def const4_rule(model, k):
    return sum(model.delta[i,j] for i in model.Iset for j in model.Jset if i+j==k)<=1
model.const4= pyo.Constraint(model.K2Set, rule = const4_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part B
model.const4.deactivate()
```

* #### Part C

```python
def const5_rule(model, i , j):
    if i<=6 and j<=7:
        return model.delta[i,j]+model.delta[i+2,j+1]<=1
    else:
        return pyo.Constraint.Skip
model.const5= pyo.Constraint(model.Iset, model.Jset , rule=const5_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const5.deactivate()
```

```python
def const6_rule(model, i , j):
    if i<=6 and j>=2:
        return model.delta[i,j]+model.delta[i+2,j-1]<=1
    else:
        return pyo.Constraint.Skip
model.const6= pyo.Constraint(model.Iset, model.Jset , rule=const6_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const6.deactivate()
```

```python
def const7_rule(model, i , j):
    if i>=3 and j<=7:
        return model.delta[i,j]+model.delta[i-2,j+1]<=1
    else:
        return pyo.Constraint.Skip
model.const7= pyo.Constraint(model.Iset, model.Jset , rule=const7_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const7.deactivate()
```

```python
def const8_rule(model, i , j):
    if i>=3 and j>=2:
        return model.delta[i,j]+model.delta[i-2,j-1]<=1
    else:
        return pyo.Constraint.Skip
model.const8= pyo.Constraint(model.Iset, model.Jset , rule=const8_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const8.deactivate()
```

```python
def const9_rule(model, i , j):
    if i<=7 and j<=6:
        return model.delta[i,j]+model.delta[i+1,j+2]<=1
    else:
        return pyo.Constraint.Skip
model.const9= pyo.Constraint(model.Iset, model.Jset , rule=const9_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const9.deactivate()
```

```python
def const10_rule(model, i , j):
    if i<=7 and j>=3:
        return model.delta[i,j]+model.delta[i+1,j-2]<=1
    else:
        return pyo.Constraint.Skip
model.const10= pyo.Constraint(model.Iset, model.Jset , rule=const10_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const10.deactivate()
```

```python
def const11_rule(model, i , j):
    if i>=2 and j<=6:
        return model.delta[i,j]+model.delta[i-1,j+2]<=1
    else:
        return pyo.Constraint.Skip
model.const11= pyo.Constraint(model.Iset, model.Jset , rule=const11_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const11.deactivate()
```

```python
def const12_rule(model, i , j):
    if i>=2 and j>=3:
        return model.delta[i,j]+model.delta[i-1,j-2]<=1
    else:
        return pyo.Constraint.Skip
model.const12= pyo.Constraint(model.Iset, model.Jset , rule=const12_rule)

#Deactivating the constraint, We activate it when we want to solve the model for part C
model.const12.deactivate()
```

* #### Solving the model for part A

```python
result=pyo.SolverFactory('gams').solve(model, solver='cplex', keepfiles=True, tee=True,io_options={'solvelink':0} , add_options=['option optcr=0;'])
```

```python
for i in model.Iset:
    for j in model.Jset:
        if pyo.value(model.delta[i,j])==1:
            print(i,j)
```

* #### Solving the model for part B

```python
#Fisrt, we deactivate the constraints in part A, then we activate the constraints for part B

model.const1.deactivate()
model.const2.deactivate()

model.const3.activate()
model.const4.activate()
```

```python
result=pyo.SolverFactory('gams').solve(model, solver='cplex', keepfiles=True, tee=True,io_options={'solvelink':0} , add_options=['option optcr=0;'])
```
